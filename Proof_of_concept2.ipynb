```python
import pandas as pd
from pathlib import Path
import numpy as np




pd.options.display.float_format = '{:,.0f}'.format
```

```python
# path to raw data

csv_path = Path('C:/Data from D drive/Temp/mls77007_raw.csv')
```

```python
mls_df = pd.read_csv(csv_path)
mls_df.head()
```

```python
# format List Price column


mls_df['List Price'] = mls_df['List Price'].str.replace('$', '')
mls_df['List Price'] = mls_df['List Price'].str.replace(',', '')
mls_df['List Price'] = mls_df['List Price'].astype(float)
```

```python
mls_df.columns
```

```python
mls_df['Rent_Sale'] = np.where(mls_df['Property Type']=='Rental', 'Rental', 'Sale')
mls_df.fillna(0, inplace=True)
mls_df.head()
```

```python
 # sqr footage bins
bins = [448, 1163,1878,2593,3308,4023,4738,5453,6168,6883, 18322]


group_names = ["448-1162", "1163-1877", "1878-2592", "2593-3307", "3308-4022", "4023-4737", "4738-5452", "5453-6167", "6168-6882","6883 - "]
```

```python
mls_df["SqrFt_Bin"] = pd.cut(mls_df["Sq Ft Total"], bins, labels=group_names)

mls_df.head()
```

```python
 # year_built bins
bins2 = [1870,1889, 1908, 1928, 1947, 1966, 1985, 2005, 2019]


group_names2 = ["1870-1888", "1889-1907", "1908-1927", "1928-1946" , "1947-1965" , "1966-1984" , "1985-2004" , "2005-2019"]

```

```python
mls_df["YrBuilt_Bin"] = pd.cut(mls_df["Year Built"], bins2, labels=group_names2)

mls_df
```

```python
 # #of bedrooms
bins3 = [0, 1, 3, 4,5]

group_names3 = ["1", "2 or 3","4", "5"]
```

```python

# add column with #bedroom bins

mls_df["numberOfBedrooms"] = pd.cut(mls_df["Bedrooms"], bins3, labels=group_names3)

mls_df.head()
```
