```python
import pandas as pd
from pathlib import Path
import numpy as np
pd.options.display.float_format = '{:,.0f}'.format
```

```python
df_raw=pd.read_csv("Resources/mls77007_csv.csv",header=1)
```

```python
df_raw.head()
```

```python
df_raw.dropna(axis=1, how="all", inplace=True)
```

```python
df_raw["List Price"]=df_raw["List Price"].str.replace("$","").astype(float)
```

```python
df_raw.head()
```

```python
df_raw.columns
```

```python
#filter on rental proerties
rental_df = df_raw[df_raw["Property Type"]=="Rental"]
df = df_raw[df_raw["Property Type"]!="Rental"]
```

```python
df.head()
```

```python
#Cost Assumptions

#Downpayment fixed at 20% for all properties
downpayment = 0.2

#Rent per square foot average taken from the current "for rent" listed properties.
avg_rent_per_sqrft = 1.56

#Vacancy Rate is expressed in percentage.
vacancy_rate = 0.05

#Property Tax for zipcode 77007 expressed in percentage.
property_tax = 0.027

#Property management fee expressed as a percentage of the rental income.
management_cost = 0.08

#Annual Repair and Maintenance Costs expressed as a percentage of the list price.
maintenance_cost = 0.01

#Loan amount expressed as a percentage of the list price
loan_amount = 0.8

#Closing cost expressed as a percentage of the loan amount
closing_cost = 0.0375

#Annual mortgage payment, estimated as a percentage of the loan amount.
mortgage_cost = 0.006*12

#Homeowner Insurance expressed as a percentage of the list price.
insurance_cost = 0.005

#Annual Financial Expenses = Mortgage Cost + Insurance Cost + Vacancy Cost

#Annual Operational Cost = Property Tax + Maintenance Cost + Property Management Cost
```

```python
df['Downpayment'] = df['List Price'] * downpayment
```

```python
df['Rent from avg'] = df['Sq Ft Total'] * avg_rent_per_sqrft
```

```python
df['Annual Income'] = df['Rent from avg'] * 12
```

```python
df['Vacancy Cost'] = df['Annual Income'] * vacancy_rate
```

```python
df['Property Tax'] = df['List Price'] * property_tax
```

```python
df['Property Management Cost'] = df['Annual Income'] * management_cost
```

```python
df['Maintenance Cost'] = df['List Price'] * maintenance_cost
```

```python
df['Loan Amount'] = df['List Price'] * loan_amount
```

```python
df['Closing Cost'] = df['Loan Amount'] * closing_cost
```

```python
df['Mortgage Cost'] = df['Loan Amount'] * mortgage_cost
```

```python
df['Insurance Cost'] = df['List Price'] * insurance_cost
```

```python
df['Annual Financial Expenses'] = df['Mortgage Cost'] + df['Insurance Cost'] + df['Vacancy Cost']
```

```python
df['Annual Operational Expenses'] = df['Property Tax'] + df['Maintenance Cost'] + df['Property Management Cost']
```

```python
df['Annual Cash Flow'] = df['Annual Income'] - df['Annual Financial Expenses'] - df['Annual Operational Expenses']
```

```python
df['Initial Investment'] = df['Downpayment'] + df['Closing Cost']
```

```python
df['Cap Rates'] = df['Annual Cash Flow']/df['Initial Investment']
```
