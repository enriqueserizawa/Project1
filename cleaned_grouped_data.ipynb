```python
import pandas as pd
df=pd.read_csv("Resources/mls77007_csv.csv",header=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(axis=1, how="all", inplace=True)
df.head()
```

```python
df.columns
```

```python
df["List Price"]=df["List Price"].str.replace("$","").astype(float)
```

```python
df.head()
```

```python
#Sq Ft Total, Year Built, Bedrooms
```

```python
#filter on rental proerties
rental_df = df[df["Property Type"]=="Rental"]
none_rental_df = df[df["Property Type"]!="Rental"]
```

```python
cols= ["Sq Ft Total", "Year Built", "Bedrooms", "Subdivision","List Price"]
none_rental_df= none_rental_df[cols].copy()
```

```python
#DATA~~~~~NONE rental properties
```

```python
none_rental_df.head()
```

```python
#new_df.info()
none_rental_df.dropna(inplace=True)
```

```python
none_rental_df.info()
```

```python
none_rental_df["Year Built"].astype(int)
```

```python
none_rental_df["Year Built"] = none_rental_df["Year Built"].astype(int)
```

```python
none_rental_df.head()
```

```python
#pd.cut(new_df["Sq Ft Total"], bins=10)
none_rental_df["sqftbins"]=pd.cut(none_rental_df["Sq Ft Total"], bins=10)
```

```python
none_rental_df["yearbuilt_bins"]=pd.cut(none_rental_df["Year Built"], bins=10)
```

```python
none_rental_df.head()
```

```python
none_rental_df.groupby(["Subdivision","sqftbins","yearbuilt_bins"]).mean().dropna()
```
